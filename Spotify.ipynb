```python
# First of all, I am importing some important libraries that I will need for this project . 

import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://chartmasters.org/most-streamed-artists-ever-on-spotify/' 
# ( This is the url link of website that I am using for this project. )
```

```python
response = requests.get(url)  # Here, we requested the webiste to fetch the data. 
```

```python
response # Here, we got the response of the website and we checked if it is >=200
```

```python
soup = BeautifulSoup(response.content,'html.parser')
# parsing the response by using html.parser
```

```python
print(soup)
```

```python
# As there are multiple 'td' tags we can use select method and than use square bracket to get the respective 'td' tag data, and than use .text method to extract the text data from the 'td' tag.
d1=soup.select('td')[0].text
d2=soup.select('td')[1].image
d3=soup.select('td')[2].text
d4=soup.select('td')[3].text
d5=soup.select('td')[4].text
d6=soup.select('td')[5].text
d7=soup.select('td')[6].text
d8=soup.select('td')[7].text
d9=soup.select('td')[8].text
d10=soup.select('td')[9].text



d1,d2,d3,d4,d5,d6,d7,d8,d9,d10
```

```python
data_iterator=soup.findAll('td')
data_iterator
```

```python
len(data_iterator)
```

```python
data=[]
```

```python
for i in range(0,len(data_iterator),10):
        Sr_no = soup.select('td')[i].text
        image = soup.select('td')[i+1].image
        Artist = soup.select('td')[i+2].text
        Lead_Streams = soup.select('td')[i+3].text
        Tracks = soup.select('td')[i+4].text
        B1 = soup.select('td')[i+5].text
        M100 = soup.select('td')[i+6].text
        M10 = soup.select('td')[i+7].text
        M1 = soup.select('td')[i+8].text
        Feat_streams = soup.select('td')[i+9].text
        data.append((
            Sr_no,
            image,
            Artist,
            int(Lead_Streams.replace(',','')),
            Tracks,
            B1,
            M100,
            M10,
            M1,
            int(Feat_streams.replace(',',''))
        ))
```

```python
data 
```

```python
df = pd.DataFrame(data,columns=['Sr_no','image','Artist','Lead_Streams','Tracks','B1',
'M100','M10','M1','Feat_Strams'])  # Naming the dataframe columns and creating the dataframe 
```
